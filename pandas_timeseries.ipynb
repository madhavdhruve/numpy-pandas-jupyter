```python
#US holiday calendar
from pandas.tseries.holiday import USFederalHolidayCalendar
from pandas.tseries.offsets import CustomBusinessDay

usb = CustomBusinessDay(calendar=USFederalHolidayCalendar())
usb
```

```python
pd.date_range(start="1/1/2017", periods=72, freq=usb) #exclude all US Holidays
```

```python
#https://github.com/pandas-dev/pandas/blob/master/pandas/tseries/holiday.py
#create custom holiday calendar
from pandas.tseries.holiday import AbstractHolidayCalendar, nearest_workday, Holiday
class myBirthdayCalendar(AbstractHolidayCalendar):
    """
    US Federal Government Holiday Calendar based on rules specified by:
    https://www.opm.gov/policy-data-oversight/
       snow-dismissal-procedures/federal-holidays/
    """
    rules = [
        Holiday('Madhav"s Birth Day', month=5, day=16, observance=nearest_workday), #here near_workday will assume near working day to be also holiday
    ]

myc = CustomBusinessDay(calendar=myBirthdayCalendar())
myc
```

```python
pd.date_range(start="4/1/2017", end="4/30/2017", freq=myc)
```
