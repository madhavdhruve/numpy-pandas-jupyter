```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")

plt.legend()
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")

plt.legend(loc="upper right")
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")

plt.legend(loc="best") #best smartly finds blank area itself and puts box there
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")

plt.legend(loc="best", shadow=True)
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")

plt.legend(loc="best", shadow=True, fontsize="large")
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")

plt.legend(loc="best", shadow=True, fontsize="large")

plt.grid()
```

```python
company = ["GOOGLE","AMAZON","MICROSOFT","FACEBOOK"]
revenue = [90,136,89,27]
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(company, revenue)
```

```python
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos,revenue, label="Revenue")
plt.legend()
```

```python
profit=[40,2,34,12]
```

```python
plt.xticks(xpos,company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos,revenue, label="Revenue")
plt.bar(xpos,profit,label="Profit")
plt.legend()
```

```python
plt.xticks(xpos,company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2,revenue, label="Revenue")
plt.bar(xpos+0.2,profit,label="Profit")
plt.legend()
```

```python
plt.xticks(xpos,company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")
plt.legend()
```

```python
#Horizontal Chart
plt.yticks(xpos,company)
plt.title("US Tech Stocks")


plt.barh(xpos-0.2,revenue,label="Revenue")
plt.barh(xpos+0.2,profit,label="Profit")
plt.legend()
```

```python
#Histograms
#They are Bar Charts that contain frequency as y-variable
blood_sugar=[113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar) #By default, it will plot 10 beans for you
```

```python
plt.hist(blood_sugar, bins=3, rwidth=0.95)
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95)
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color="g", histtype="step") #color="green" also works
```

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.xlabel("sugar range")
plt.ylabel("Total no. of patients")
plt.title("Blood Sugar Analysis")

plt.hist(
    [blood_sugar_men, blood_sugar_women],
    color=["green", "orange"],
    bins=[80,100,125,150],
    rwidth=0.95, label=["men", "women"],
    orientation="horizontal"
)
plt.legend()
```

```python
#Pie Chart
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utilities"]

plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.2f%%")
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.0f%%", shadow=True)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.0f%%", shadow=True, explode=[0,0.2,0,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.0f%%", shadow=True, explode=[0,0.2,0.2,0.2,0.2])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.0f%%", shadow=True, explode=[0,0.2,0.2,0.2,0.2], startangle=45)
plt.show()
```

```python
#save chart to file
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct="%0.0f%%", shadow=True, explode=[0,0.2,0.2,0.2,0.2], startangle=45)
plt.savefig("piechart.png",bbox_inches="tight", pad_inches=2, transparent=True) #bbox_inches will ajust image properly and not cut any of its parts
#pad_inches adds padding in image file
#transparent adds transparent color in background
```
