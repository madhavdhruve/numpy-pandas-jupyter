```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","bangalore"],
    "temperature": [32,45,30],
})
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai","bangalore"],
    "windpeed": [7,12,9],
})
pd.merge(temperature_df,windspeed_df,on="city") #merge automatically handles mis-indexed data instead of concat expecting it to give individual indexes to display correctly
```

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","ahmedabad"],
    "temperature": [32,45,30],
})
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai","pune"],
    "windpeed": [7,12,9],
})
pd.merge(temperature_df,windspeed_df,on="city") 
```

```python
print(pd.merge(temperature_df,windspeed_df,on="city",how="outer"))  #Outer Join according to set therory
print(pd.merge(temperature_df,windspeed_df,on="city",how="inner"))  #Inner Join according to set therory
print(pd.merge(temperature_df,windspeed_df,on="city",how="left"))  #Left Join according to set therory
print(pd.merge(temperature_df,windspeed_df,on="city",how="right"))  #Right Join according to set therory
print(pd.merge(temperature_df,windspeed_df,on="city",how="right", indicator=True))  #indicator indicates where that value came from
```

```python
import csv
from datetime import date,timedelta
from random import randint, uniform
ls = ["Date","Facebook","google","Microsoft","Facebook","google","Microsoft"]
start=date(2017,6,5)
ls
lsd = []
for i in range(5000):
    t=[]
    t.append(start.strftime("%Y-%m-%d"))
    start+=timedelta(1)
    t.append(str(randint(0,50)))
    t.append(str(randint(0,50)))
    t.append(str(randint(0,50)))
    t.append(str(round(uniform(0,50),2)))
    t.append(str(round(uniform(0,50),2)))
    t.append(str(round(uniform(0,50),2)))
    lsd.append(t)
lsd
with open("abc.csv", "w") as f:
    f.write(",".join(ls)+"\n")
    for l in lsd:
        f.write(",".join(l)+"\n")
```
